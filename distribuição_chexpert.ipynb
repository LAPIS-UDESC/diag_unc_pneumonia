```python
import pandas
from sklearn.utils import resample
import os

DATASET_PATH = os.path.join(os.getcwd(), 'chexpert/chexpertchestxrays-u20210408')
TRAIN_CSV_PATH = os.path.join(DATASET_PATH, 'train_cheXbert.csv')

df = pandas.read_csv(TRAIN_CSV_PATH)
df = df[df['Frontal/Lateral'] == 'Frontal']

# Substituindo NaN por zero
# df.fillna(0, inplace=True)

# Casos que possuem pneumonia positiva
df_pneumonia = df[df['Pneumonia'] == 1]

# Casos que possuem pneumonia incerta
df_unc_pneumonia = df[df['Pneumonia'] == -1]

# Casos que possuem pneumonia negativa, mas possuem outras morbidades
df_no_pneumonia = df[df['Pneumonia'] == 0]

df_list = [df_pneumonia, df_unc_pneumonia, df_no_pneumonia]
```

```python
import matplotlib.pyplot as plt

plt.bar(
    ['Pneumonia', 'Uncertain Pneumonia', 'No Pneumonia'],
    [len(df) for df in df_list],
)

plt.xlabel('Class')
plt.ylabel('Num. of Cases')
plt.title('Class Sizes')
```
